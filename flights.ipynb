>## The flights dataset from Udacity data analysis Nanodegree 

```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pd.set_option('display.max_columns', None) #show all headers

#importing the datasets
flights = pd.read_csv('flights.csv')
airports = pd.read_csv('airports.csv')
flights.head()
```

```python
airports.head()
```

```python
#join the two tables 
df = flights.merge(airports, left_on='DESTINATION_AIRPORT', right_on='IATA_CODE')
```

```python
df.head(3)
```

>#### Number Of Flights in the dataset

```python
number_of_flights = df.shape[0]
number_of_flights
```

>#### Number Of Completed Flights in the dataset

```python
number_of_completed_flights = df.query('CANCELLED == 0').shape[0]
print('number of flights= ',number_of_completed_flights)
```

```python
df.CANCELLATION_REASON.value_counts()
```

```python
#replace the letters with the real reason for cancelation
df['CANCELLATION_REASON'].replace(['A','B','C'],['Airline/Carrier','Weather','National Air System'],inplace=True)
df.head()
```

>#### Number canceled flights due to each reason

```python
df.CANCELLATION_REASON.value_counts()
```

```python
#replace the NaN values with Completed
df['CANCELLATION_REASON'].fillna('Completed',inplace=True)
df.head(3)
```

```python
df.CANCELLATION_REASON.value_counts() 
```

>#### Number Of Flights by City (TOP 10 cities)

```python
flights_by_city_10 = df.CITY.value_counts()[:10]
flights_by_city_10
```

```python
# create the horizontal bar chart
flights_by_city_10.sort_values().plot(kind='barh')

# add data labels to the chart
for i, v in enumerate(flights_by_city_10.sort_values()):
    plt.text(v + 0.5, i, str(v), color='black')

# add axis labels and title
plt.xlabel('Number of flights')
plt.ylabel('City')
plt.title('Number of Flights by CITY TOP10');
```

>#### Number Of Flights by Month

```python
flights_by_month = df.MONTH.value_counts().sort_index()
flights_by_month
```

```python
# create the line chart
flights_by_month.plot(kind='line')

# add data labels to the chart
for i, v in enumerate(flights_by_month):
    plt.text(i + 0.1, v + 100, str(v), color='black')

# add axis labels and title
plt.xlabel('Month')
plt.ylabel('Number of flights')
plt.title('Number of Flights by Month');
```

>#### Average arrival delay time for all cities

```python
df.ARRIVAL_DELAY.mean()
```

>#### Average arrival delay time for the top 10 cities in number of flights

```python
flights_by_city_10
```

```python
# group the original dataframe by 'CITY' column and calculate the mean of 'ARRIVAL_DELAY' for each city
city_arrival_delay_means = df.groupby('CITY')['ARRIVAL_DELAY'].mean().round(2)

# filter the data to keep only the cities in flights_by_city_10
city_arrival_delay_means_10 = city_arrival_delay_means[flights_by_city_10.index].sort_values(ascending=False)
city_arrival_delay_means_10
```

```python
# create the horizontal bar chart
city_arrival_delay_means_10.sort_values().plot(kind='barh')

# add data labels to the chart
for i, v in enumerate(city_arrival_delay_means_10.sort_values()):
    plt.text(v + 0.1, i, str(v), color='black')

# add axis labels and title
plt.xlabel('Average arrival delay')
plt.ylabel('City')
plt.title('Average arrival delayY by CITY TOP10');
```

>#### Average arrival delay time by month

```python
average_arrival_delay_by_month = df.groupby('MONTH')['ARRIVAL_DELAY'].mean().round(2)
average_arrival_delay_by_month
```

```python
# create the line chart
average_arrival_delay_by_month.plot(kind='line')

# add data labels to the chart
for i, v in enumerate(average_arrival_delay_by_month):
    plt.text(i + 1, v + .1, str(v), color='black')
    
# add axis labels and title
plt.xlabel('Month')
plt.ylabel('Average arrival delay')
plt.title('Average arrival delay by Month');
```
